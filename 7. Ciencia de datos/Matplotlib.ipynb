```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 2, 3, 4, 5])
```

```python
#Gráfico de líneas

plt.title("Gráfico de líneas")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")

plt.plot(x,y, color="green", marker="8", markersize = 5)
plt.savefig("lineas.pdf")

plt.show()
```

```python
#Gráfico de dispersión

plt.title("Gráfica de dispersión")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")

plt.scatter(x,y, color="blue")
plt.grid()
plt.show()
```

```python
#Gráfico de barras vertical

plt.title("Gráfica de barras")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")

x = ["A", "B", "C", "D", "E"]


plt.bar(x,y, color="yellow")
plt.savefig("barras.png")

plt.show()
```

```python
#Gráfico de barras horizontal

plt.title("Gráfica de barras horizontal")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")

x = ["A", "B", "C", "D", "E"]


plt.barh(x,y, color="red")
plt.show()
```

```python
#Gráfico de pastel

plt.title("Gráfico de pastel")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")

plt.pie(y)
plt.show()

```

```python
#Histograma

plt.title("Histograma")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")


x = np.random.normal(100, 30, 200)
# np.random.normal(media, desviacion estandar, población)


plt.hist(x, color="orange")
# plt.show()
```

```python
#Gráfico de pasos

x = np.array([1, 2, 3, 4, 5])
y = np.array([6, 7, 8, 4, 10])

plt.title("Gráfica de pasos")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")

plt.step(x, y, color="black")
plt.show()
```
