```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
datosX = np.array([1,2,3,4,5,6])
datosY = np.array([1,2,3,4,5,6])

plt.title("Gráfico de líneas")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.plot(datosX, datosY, color="green", marker="D", markersize = 8, linestyle="--", label="Gráfica verde")
plt.legend()
plt.grid()
plt.show()
```

```python
#Subplots

fig, ax = plt.subplots(nrows=2, ncols=2)
ax[0,0].bar(datosX,datosY, color="yellow")
ax[1,0].scatter(datosX, datosY, color="blue", label="Gráfica azul")
ax[1,1].plot(datosX, datosY, color="green", marker="D", markersize = 8, linestyle="--", label="Gráfica verde")
plt.show()
```

```python
#Subplots con nombres personalizados

fig, ((graficoBarra, graficoVacio), (graficoDispersion, graficoLinea)) = plt.subplots(nrows=2, ncols=2)
graficoBarra.bar(datosX,datosY, color="orange")
graficoDispersion.scatter(datosX, datosY, color="blue", label="Gráfica azul")
graficoLinea.plot(datosX, datosY, color="green", marker="D", markersize = 8, linestyle="--", label="Gráfica verde")
plt.show()
```

```python
#Subplots con nombres personalizados 2

#Nota: Cuando personalizamos el nombre de los axes (gráficos contenidos dentro del figure) nosotros debemos
# agrupar tuplas igual al número de filas que tengamos y cada tupla debe contener el número de columnas dadas

fig, ((graficoBarra, graficoVacio), (graficoDispersion, graficoLinea), (graficoBarraH, graficoPastel)) = plt.subplots(nrows=3, ncols=2)
graficoBarra.bar(datosX,datosY, color="orange")
graficoDispersion.scatter(datosX, datosY, color="blue", label="Gráfica azul")
graficoLinea.plot(datosX, datosY, color="green", marker="D", markersize = 8, linestyle="--", label="Gráfica verde")
graficoBarraH.barh(datosX,datosY, color="cyan")
graficoPastel.pie(datosX)
plt.show()
```

```python
#Subplots con uno de los plots formado de manera simultanea

datosZ = [6,5,4,3,2,1]

fig, ((graficoBarra, graficoSimultaneo), (graficoFill, graficoLinea), (graficoBarraH, graficoPastel)) = plt.subplots(nrows=3, ncols=2, figsize=(20, 20))
#figsize solicita ancho y alto en pulgadas
graficoBarra.bar(datosX,datosY, color="orange")
graficoBarra.set_title("grafico Barra")
graficoBarra.set_xlabel("Eje X")
graficoBarra.set_ylabel("Eje Y")

graficoSimultaneo.plot(datosX, datosY, color = "red", label="linea")
graficoSimultaneo.scatter(datosX, datosZ, color = "black", label="puntos")
graficoSimultaneo.legend()
graficoFill.fill_between(datosX, datosY, color="blue", label="Gráfica azul")
graficoLinea.plot(datosX, datosY, color="green", marker="D", markersize = 8, linestyle="--", label="Gráfica verde")
graficoBarraH.barh(datosX,datosY, color="cyan")
graficoPastel.pie(datosX)


fig.savefig("contenedor.png")
plt.show()
```
