```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arregloX = [1,2,3,4,5]
arregloY = [1,2,3,4,5]
arregloY2 = [0,0,0,0,0]
plt.plot(arregloX, arregloY, "-o", color="red")
plt.plot(arregloX, arregloY2, "-o", color="green")
plt.fill_between(arregloX, arregloY, arregloY2)
```
